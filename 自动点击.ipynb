```python
import glob
def get_real_time(v_str):
    total_seconds= int(v_str.split('.')[0])
    #second = total_seconds % 60
    #hour= total_seconds // 60 // 60
    #minute = total_seconds // 60 % 60
    #ret_time = str(hour)+"时"+str(minute)+"分"+str(second)+"秒"
    return total_seconds
```

```python
from bs4 import BeautifulSoup as BS
import time
files = sorted(glob.glob(r"C:\Users\CANDY\Desktop\毕设\弹幕\papi\752925271.xml"))
for index, filename in enumerate(files, 1):
    file = open(filename, encoding='utf-8').read()
    soup = BS(file, 'xml')
danmu_list=soup.find_all('d')
#print('共爬取到{}条弹幕'.format(len(danmu_list)))
```

```python
import pandas as pd
import numpy as np
data =pd.DataFrame([])
```

```python
for d in danmu_list:
    time=get_real_time(d['p'])
    content=d.text
    info={"时间s":time,"弹幕内容":content}
    data = data.append(info,ignore_index=True)
```

```python
data.sort_values(by="时间s" , inplace=True, ascending=True) 
data=data.reset_index(drop=True)
```

```python
for i in range(len(data)):
    if ('猝不及防' in data["弹幕内容"][i])or('快跑' in data["弹幕内容"][i])or ('图穷匕见' in data["弹幕内容"][i])or('我就知道' in data["弹幕内容"][i])or('味道不对' in data["弹幕内容"][i]):
        print(data["时间s"][i],data["弹幕内容"][i])
```
